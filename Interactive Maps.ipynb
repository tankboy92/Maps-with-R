```r
library(readr)
library(tidyverse)
library(maps)
library(mapproj)
```

```r
wage <- read.csv('state_minimum_wage_data.csv')
```

```r
wage.2020 <- subset(wage, Year == 2020)
```

```r
wage.2020
```

```r
# Create a map of the US with the minimum wage for each state in 2020

# Import Packages

install.packages("maps")
install.packages("mapproj")

library(readr)
library(tidyverse)
library(maps)
library(mapproj)
```

```r
us_states<- map_data("state")
```

```r
us_states
```

```r
# Capitalize the state names in us_states

us_states <- rename(us_states, State = region)

```

```r
library(tools)

us_states$State <- toTitleCase(us_states$State)
```

```r
us_states
```

```r
merged.data <- merge(us_states, wage.2020, by="State")
```

```r
# Plot the map

m1 <- ggplot(data=merged.data, aes(x=long, y=lat, group=State, fill=State.Minimum.Wage)) +
  geom_polygon(color="black", linewidth=0.1) +
  theme_void() +
coord_map("albers", lat0=39, lat1=45)
```

```r
m1
```

```r
install.packages("ggiraph")
```

```r
library(ggiraph)
```

```r
# Make Interactive Map

m2 <- ggplot(data=merged.data, aes(x=long, y=lat, group = State)) +
  geom_polygon_interactive(aes(tooltip=State.Minimum.Wage, data_id = State), color="black", linewidth=0.1,hover_nearest=TRUE) +
  theme_void() +
  coord_map("albers", lat0=39, lat1=45)

```

```r
m2
```

```r
m3 <- girafe(ggobj = m2)
```

```r
m3
```

```r
merged.data.2 <- left_join(us_states, wage.2017, by="State")
```
